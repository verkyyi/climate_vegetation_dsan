```python
## For historic data, combine the rows with the same year, long and lat
# for each feature, if all the values are the same, keep the value, otherwise use the mean of non-NaN values
import numpy as np
# Custom aggregation function
def custom_agg(x):
    if x.dtype != 'object': # if the column is numeric
        if x.isnull().all(): # if all the values are NaN
            return np.nan
        return x.mean() # return the mean of non-NaN values
    else: # if the column is not numeric
        return x.mode().iloc[0]

# Apply the custom aggregation function
historic_data_grouped = historic_data.groupby(['year', 'long', 'lat']).agg(custom_agg).reset_index()
historic_data_grouped
```

```python
# prepare export dataset from the grouped data
base_columns = ['long', 'lat', 'year']
climate_columns = ['T_Summer', 'PPT_Summer', 'VWC_Summer_whole', 'ExtremeShortTermDryStress_Summer_whole']
vegetation_columns = ['Herb', 'Litter', 'Bare', 'Shrub', 'Ann_Herb', 'treecanopy']
export_dataset = historic_data_grouped[base_columns + climate_columns + vegetation_columns]
# order by 'long', 'lat', 'year'
export_dataset = export_dataset.dropna()
export_dataset = export_dataset.sort_values(['long', 'lat', 'year']).reset_index(drop=True)
export_dataset.to_csv('./datasets/cleaned/export_dataset.csv', index=False)
export_dataset
```

```python
# get average values across all years for each location
average_per_locations = export_dataset.groupby(['long', 'lat']).mean().reset_index()
# drop the year column
average_per_locations = average_per_locations.drop(columns=['year']+vegetation_columns)
# keep only 2 decimal places for feature values(not for long and lat)
# convert the feature values into [0, 1] range
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# save converted values into columns with the same name but with a suffix '_t'
climate_columns_t = [col+'_t' for col in climate_columns]
average_per_locations[climate_columns_t] = scaler.fit_transform(average_per_locations[climate_columns])
average_per_locations = average_per_locations.apply(lambda x: round(x, 2) if x.name not in ['long', 'lat'] else x)
average_per_locations.to_csv('./datasets/cleaned/average_per_locations.csv', index=False)
average_per_locations
```

```python
# get points for the year of 1980
year_1980 = export_dataset[export_dataset['year'] == 1980]
year_1980 = year_1980.drop(columns=['year'] + vegetation_columns)
# keep only 2 decimal places for feature values(not for long and lat)
# convert the feature values into [0, 1] range
year_1980[climate_columns_t] = scaler.fit_transform(year_1980[climate_columns])
year_1980 = year_1980.apply(lambda x: round(x, 2) if x.name not in ['long', 'lat'] else x)
year_1980.to_csv('./datasets/cleaned/year_1980.csv', index=False)
year_1980
```

```python
# get points for the year of 2018 
year_2018 = export_dataset[export_dataset['year'] == 2018]
year_2018 = year_2018.drop(columns=['year']+ vegetation_columns)
# keep only 2 decimal places for feature values(not for long and lat)
# convert the feature values into [0, 1] range
year_2018[climate_columns_t] = scaler.fit_transform(year_2018[climate_columns])
year_2018 = year_2018.apply(lambda x: round(x, 2) if x.name not in ['long', 'lat'] else x)
year_2018.to_csv('./datasets/cleaned/year_2018.csv', index=False)
year_2018
```

```python
# get comparison points for the year of 1980 and 2018, excluding those _t columns
comparison = pd.merge(year_1980, year_2018, on=['long', 'lat'], suffixes=('_1980', '_2018'))
comparison
# filter out those columns with _t_1980 and _t_2018
comparison = comparison[[col for col in comparison.columns if '_t' not in col]]
comparison
# calculate the difference between 2018 and 1980
for col in climate_columns:
    comparison[col+'_diff'] = comparison[col+'_2018'] - comparison[col+'_1980']
comparison
# round the values to 2 decimal places
comparison = comparison.apply(lambda x: round(x, 2) if x.name not in ['long', 'lat'] else x)
# save the comparison dataset
comparison.to_csv('./datasets/cleaned/comparison.csv', index=False)
comparison
```

```python
# get vegetation points for each location
vegetation_points = export_dataset[['long', 'lat'] + vegetation_columns].drop_duplicates().reset_index(drop=True)
# convert the vegetation values into [0, 1] range
vegetation_columns_t = [col+'_t' for col in vegetation_columns]
vegetation_points[vegetation_columns_t] = scaler.fit_transform(vegetation_points[vegetation_columns])
# keep only 2 decimal places for vegetation values
vegetation_points = vegetation_points.apply(lambda x: round(x, 2) if x.name not in ['long', 'lat'] else x)
vegetation_points.to_csv('./datasets/cleaned/vegetation_points.csv', index=False)
vegetation_points
```
